```python
import pandas as pd
import yfinance as yf
```

```python
sp_wiki_url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
sp_wiki_df_list = pd.read_html(sp_wiki_url)
```

```python
sp_wiki_df_list
```

```python
len(sp_wiki_df_list)
```

```python
sp_df = sp_wiki_df_list[0]
```

```python
sp_df.head()
```

```python
sp_df['Symbol'].values
```

```python
sp_ticker_list = list(sp_df['Symbol'].values)
```

```python
sp_ticker_list
```

```python
df = yf.download(sp_ticker_list)
```

```python
adj_close = df["Adj Close"]
```

```python
adj_close.plot()
```

```python
adj_close.columns
```

```python
adj_close[['A', 'AAL', 'AAPL', 'ABBV', 'ABNB', 'ABT']].plot(subplots=True)
```

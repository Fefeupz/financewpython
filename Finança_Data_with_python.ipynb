```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import yfinance as yf 
```

```python
%matplotlib inline 
```

```python
stocks = dict(
    AAPL = 'Apple Stock',
    AMZN = 'Amazon Stock',
    CNY = 'CNY/USD Exchange Rate',
    GDX = 'Gold Price',
    GLD = 'SPDR Gold Trust',
    GS = 'Goldman Sachs Stock',
    INTC = 'Intel Stock',
    MSFT = 'Microsoft Stock',
    SPY = 'S&P 500')
```

```python
stocks.keys()
```

```python
list(stocks.keys())
```

```python
df = yf.download(list(stocks.keys()))
```

```python
df.tail()
```

```python
adj_close = df['Adj Close']
```

```python
adj_close
```

```python
adj_close.dropna().plot(subplots=True, figsize=(12, 10));
```

```python
adj_close = adj_close.dropna()
```

```python
for key, value in stocks.items():
    print(f'{key:8s}|{value}')
```

```python
adj_close.describe().round(2)
```

```python
adj_close.mean()
```

```python
adj_close.aggregate([min, np.mean, np.std, np.median, max]).round(2)
```

```python
adj_close.diff().head()
```

```python
adj_close.diff().mean().head()
```

```python
adj_close.pct_change().round(3).head()
```

```python
adj_close.pct_change().mean().plot(kind='bar', figsize=(6,6));
```

```python
adj_close.shift(1)
```

```python
rets = np.log(adj_close / adj_close.shift(1))
```

```python
rets.cumsum().apply(np.exp).plot(figsize=(10,5));
```
